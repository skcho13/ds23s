### `assign()`

````python
['title', 'time'] + [col for col in df.columns if col.startswith('mag')]
````

````python
cols = ["year", "month", "day", "distance", "air_time"] + \
        [col for col in flights.columns if col.endswith("delay")]  # string method .endswith
flights_sml = flights[cols].copy()
flights_sml
````

````python
# 새로 만들어진 변수는 맨 뒤로
flights_sml.assign(
    gain=lambda x:x.dep_delay - x.arr_delay,
    speed=lambda x:x.distance / x.air_time * 60
)
````

````python
# 앞에서 만든 변수나 함수를 이용할 수 있음
flights_sml.assign(
    gain=lambda x:x.dep_delay - x.arr_delay,
    hours=lambda x:x.air_time / 60,
    gain_per_hour=lambda x:x.gain / x.hours,
    rounded=lambda x: np.round(x.gain_per_hour, 1)  # use a numpy function
)
````

````python
# Find the fastest flights
(
    flights_sml
    .assign(speed=lambda x:x.distance / x.air_time)
    .sort_values(by="speed", ascending=False)
    .head(5)
)
````

### apply()
Apply: General split-apply-combine
McKineey's Chapter 10.3 참고

The most general-purpose GroupBy method is apply, which is the subject of this section.  
apply splits the object being manipulated into pieces, invokes the passed function on each piece, and then attempts to concatenate the pieces.

````python
#| layout-ncol: 2
#| output: false
tips = sns.load_dataset("tips")
tips = tips.assign(tip_pct = lambda x: x.tip / x.total_bill)
tips.head(3)
````

````python
#| echo: false
print(tips.head(3))
````

````python
def top(df, n=5, column="tip_pct"):
    return df.sort_values(column, ascending=False)[:n]
````

````python
#| layout-ncol: 2
#| output: false
top(tips, n=4)
````

````python
#| echo: false
print(top(tips, n=4))
````

````python
#| output: false
tips.groupby("time").apply(top)
````

````python
#| echo: false
print(tips.groupby("time").apply(top))
````

````python
#| output: false
tips.groupby(["time", "day"]).apply(top, n=1, column="total_bill")
````

````python
#| echo: false
print(tips.groupby(["time", "day"]).apply(top, n=1, column="total_bill"))
````
